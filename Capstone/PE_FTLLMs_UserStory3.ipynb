```python
import numpy as np
import torch
import torch.nn as nn
from transformers import Trainer
from sklearn.utils.class_weight import compute_class_weight

# Class distribution from your dataset
class_counts = np.array([7093, 20673, 17849])

# Compute class weights (inverse frequency)
class_weights = compute_class_weight(
    class_weight="balanced",
    classes=np.array([0, 1, 2]),  # Convert to NumPy array
    y=np.concatenate([[0] * class_counts[0], [1] * class_counts[1], [2] * class_counts[2]])
)

# Convert to PyTorch tensor
class_weights = torch.tensor(class_weights, dtype=torch.float).to("cuda")

# Custom Trainer with weighted loss
class CustomTrainer(Trainer):
    def compute_loss(self, model, inputs, return_outputs=False, **kwargs):  # Accept extra arguments
        labels = inputs.pop("labels")
        outputs = model(**inputs)
        logits = outputs.logits
        loss_fn = nn.CrossEntropyLoss(weight=class_weights)  # Apply weighted loss
        loss = loss_fn(logits, labels)
        return (loss, outputs) if return_outputs else loss

# Use the custom trainer
trainer = CustomTrainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset["train"],
    eval_dataset=tokenized_dataset["test"],
)

trainer.train()
```

# The above are attempts to get a good training loss and validation loss.

Upon reviewing and analysing the loss result, I came to the conclusion that the problem with the rising validation loss is due to the uneven split that makes one class underrepresented. With this fact, I made the assumption that the model was overfitting due to this uneven split. Hence, on the following attempt, I set the dataset size to the lowest of the three labels, and saw an immediate improvements to both the training and validation loss, as well as a confident accuracy result. 

```python
from datasets import load_dataset
import pandas as pd
from collections import Counter
import random

# Load the dataset
dataset = load_dataset("tweet_eval", "sentiment")

# Access the train split
train_data = dataset["train"]

# Convert to pandas DataFrame for easier manipulation
df = pd.DataFrame(train_data)

# Check the first few rows
print(df.head())

# Check class distribution
labels = df["label"]
print("Original class distribution:", Counter(labels))

# Define the target class size (7,093 samples per class)
target_class_size = 7093

# Create separate DataFrames for each class
df_negative = df[df["label"] == 0]
df_neutral = df[df["label"] == 1]
df_positive = df[df["label"] == 2]

# Randomly sample 7,093 samples from each class
df_negative_sampled = df_negative.sample(target_class_size, random_state=42)
df_neutral_sampled = df_neutral.sample(target_class_size, random_state=42)
df_positive_sampled = df_positive.sample(target_class_size, random_state=42)

# Combine the sampled data back together into a balanced dataset
balanced_df = pd.concat([df_negative_sampled, df_neutral_sampled, df_positive_sampled])

# Shuffle the balanced dataset
balanced_df = balanced_df.sample(frac=1, random_state=42).reset_index(drop=True)

# Check the new class distribution
print("New class distribution:", Counter(balanced_df["label"]))

# Preprocessing function (tokenizing)
def preprocess_function(examples):
    return tokenizer(examples['text'], truncation=True, padding='max_length', max_length=128)

# Tokenize the balanced dataset
tokenized_dataset = balanced_df.apply(lambda x: preprocess_function(x), axis=1)

# Now, you can use the tokenized dataset for your training
```
